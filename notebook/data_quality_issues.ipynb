<H2> Checking Null Values in Datasets </H>

```python
receipts.isnull().mean().sort_values(ascending=False)
```

```python
users.isnull().mean().sort_values(ascending=False)
```

```python
brands.isnull().mean().sort_values(ascending=False)
```

Noticed high missingness in all three JSON files. These should be validated or excluded from modeling.

<H2> Checking for Duplicate in Ids (PKs) in Data Model </H2>

```python
receipts['receipt_id'] = receipts['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) else None)
print(receipts.duplicated(subset='receipt_id').sum())
```

```python
brands['brand_id'] = brands['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) else None)
print(brands.duplicated(subset='brand_id').sum())
```

```python
users['user_id'] = users['_id'].apply(lambda x: x['$oid'] if isinstance(x, dict) else None)
print(users.duplicated(subset='user_id').sum())
```

Noticed some duplications of ids in user table. These should be consolidated.
